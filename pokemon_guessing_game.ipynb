```python
import requests
import random

def play_pokemon_game():
    # Pick a random ID from the original 151 (Generation 1)
    poke_id = random.randint(1, 151)
    
    url = f"https://pokeapi.co/api/v2/pokemon/{poke_id}"

    print("🔍 Fetching a random Pokémon from the database...")

    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()

        # Extracting the info we need
        name = data['name']
        types = [t['type']['name'] for t in data['types']]
        
        print("\n--- WHO'S THAT POKÉMON? ---")
        print(f"Hint: This Pokémon is a {', '.join(types)} type.")
        
        attempts = 3
        while attempts > 0:
            guess = input(f"Guess the name ({attempts} left): ").strip().lower()

            if guess == name:
                print(f"✅ Correct! It's {name.capitalize()}! 🎉")
                return
            else:
                attempts -= 1
                if attempts > 0:
                    print("❌ Not quite! Try again.")
                else:
                    print(f"💀 Game Over! The answer was {name.upper()}.")

    except Exception as e:
        print(f"Connection Error: {e}")

if __name__ == "__main__":
    play_pokemon_game()
```
